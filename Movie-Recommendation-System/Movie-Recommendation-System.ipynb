```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
def get_title(index):
    return df[df.index == index]['title'].values[0]

def get_index(title):
    return df[df.title == title].index.values[0]
```

```python
features = ['title','genres','overview']
```

```python
for feature in features:
    df[feature] = df[feature].fillna('')
```

```python
df['title'].isnull().sum()
```

```python
df['genres'].isnull().sum()
```

```python
df['combined_features'] = df.apply(lambda row: row['title'] + '|' + row['genres'] + '|' + row['overview'], axis=1)
```

```python
df[['title','genres','overview']].head()
```

```python
df.combined_features
```

```python
vectorizer = CountVectorizer()
feature_matrix = vectorizer.fit_transform(df['combined_features'])
feature_matrix
```

```python
cosine_similarities = cosine_similarity(feature_matrix)
print(cosine_similarities)
```

```python
movie_user_interests = input('Enter the Movie You Were Interested In! :')

movie_index = get_index(movie_user_interests)

similar_movies = list(enumerate(cosine_similarities[movie_index]))

sorted_similar_movies = sorted(similar_movies, key = lambda x: x[1], reverse = True)[1:]

print("\n Top 5 Similar Movies to Watch are: \n")
i = 0
for element in sorted_similar_movies:
    print(get_title(element[0]))
    i = i+1
    if i>= 5:
        break
```
